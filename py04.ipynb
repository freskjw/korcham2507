## 데이터프레임 Dataframe

```python
import pandas as pd

df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], index=["a","b","c"], columns=["A","B","C"])
df
```

```python
num = np.arange(1,10).reshape(3,3)
df = pd.DataFrame(num, index=["a","b","c"], columns=["A","B","C"])
df
```

```python
dd = {"과일이름": ["apple","banana","kiwi","melon","orange","mange"],
      "가격":[4000,5000,2500,7000,2000,8800], 
      "개수":[3,8,11,22,6,5]}
df2 = pd.DataFrame(dd)
df2.index = range(1,7)
df2.index.name = "No."
df2
```

```python
df2.columns.name = "info"
df2.columns = ['과일이름','원가','수량']
df2
```

```python
df2.describe()
```

```python
df3 = pd.DataFrame(df2, columns = ['과일이름','품종','원가','할인','수량','합계'])
df3
```

```python
df3["품종"] = df3["품종"].astype('object')
df3.loc[:,'품종'] = ['신선한','궁금한','맛있는','최적의','웰빙','유기농']
df3.loc[:,'할인'] = [0.2,0.3,0.2,0.1,0.2,0.3]
df3
```

```python
df3['합계'] = (df3["원가"] * df3["수량"] * (1 - df3["할인"])).astype('int')
df3
```

```python
df3['etc'] = "-"
df3['Test'] = "test"
print(df3.dtypes)
df3
```

```python
del df3['test']
df3
```

```python
del df3['Test']
df3
```

```python
ee = pd.Series([1.3,2.2], index=[2,4])
df3['ETC'] = ee
df3 = df3.fillna('')                                      # NaN를 내가 원하는 값으로 채우겠다.
df3
```

```python
df3['PASS'] = ''
df3
```

```python
pp = np.where(df3['합계']<=30000,'구매각','-')
df3['PASS'] = pp
df3
```

```python
# 수량이 7개 이상인 물품만, '포장' or '번들'로 표기하라.
# 컬럼 이름은 DV

df3['DV']=''
dvp = np.where(df3['수량']>7, '포장','번들')
df3['DV'] = dvp
df3
```

```python
# df3.iloc[1,1:4]
# df3.loc[2,:]
df3.loc[df3.loc[:,'과일이름']=='banana','품종':'수량']
```

```python
# 구매각인 과일만 과일이름과 품종 원가를 출력하라.

df3.loc[df3.loc[:,'PASS']=='구매각', '품종':'원가']
```

```python
# 수동으로 원하는 내용을 뽑을 수 있음
# 조건을 쓰면 자동으로 하는것

# df3.iloc[[0,2,4],[0,2,5,7]]
# df3.loc['1':'5']
df3.loc[[1,3,5],['과일이름','원가','합계','PASS']]
```

```python
import numpy as np
import pandas as pd

print("인구수 시리즈 데이터")
population_dict={
    'korea':5182,
    'japan':12622,
    'china':141178,
    'usa':32976
}
print("GDP 시리즈 데이터")
GDP_dict={
    'korea':169320000,
    'japan':516700000,
    'china':140925000,
    'usa':2041280000
}
population = pd.Series(population_dict)
GDP = pd.Series(GDP_dict)
print(population)
print(GDP)
print("="*30)
print("데이터 프레임 생성")

country = pd.DataFrame({"population":population,"GDP":GDP})
country
```

```python
country.loc['india'] = [142222,150505000]
country
```

```python
# korea만 꺼내기

country.loc[['korea'],:]
```

```python
country['GDP_per'] = country['GDP'] / country["population"]
country
```

```python
country.loc['africa'] = pd.Series([123456,7777777], index=["population","GDP"])
country
```

```python
# 결측값 확인 ,isnull() / isna()  둘이 똑같음
# 결측값 확인된걸 지우는것 .drop

country = country.drop('africa', axis=0)                                              # africa를 몽땅 날리는 것
# country.drop(country[country.isnull().any(axis=1)].index, axis=0)
country
```
